Neste exercício criaremos uma função para plotar automaticamente gráficos para a base de dados SINASC_RO 2019 com base em parêmetros diversos.

Criando imports necessários e códigos/função para plotagem de gráficos para os meses: Março, Abril, Maio, Junho e Dezembro de 2019 da base SINASC_RO

Criando plotagens com função para o mês de Março 2019

```python

#Importando as bibliotecas

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set()

#função 

def plota_pivot_table(df, value, index, func, ylabel, xlabel, opcao='nada'):
    if opcao == 'nada':
        pd.pivot_table(df, values=value, index=index, aggfunc=func).plot(figsize=(15, 5))
    elif opcao == 'sort':
        pd.pivot_table(df, values=value, index=index, aggfunc=func).sort_values(value).plot(figsize=(15, 5))
    elif opcao == 'unstack':
        pd.pivot_table(df, values=value, index=index, aggfunc=func).unstack().plot(figsize=(15, 5))
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    
    return None

#Mês de Março
sinasc = pd.read_csv('SINASC_RO_2019_MAR.csv')

#Selecionando nome do arquivo a partir dos 7 primeiros caracteres da Data de Nascimento
max_data = sinasc.DTNASC.max()[:7]
print(max_data)

# obs: 'r' é utilizado para não criar confusão com o Python em caso de \n (o r leva a interpretação literal da string)
diretorio_salvamento = r'C:\Users\giova\OneDrive\Área de Trabalho\EBAC Apostilas e Notebooks - Copia\Modulo 14'

#O método os.path.join é usado para criar o caminho completo até o diretório onde queremos salvar os arquivos.
os.makedirs(os.path.join(diretorio_salvamento, max_data), exist_ok=True)

#Criando as plotagens seguindo os parâmetros da função:
#(df, value, index, func, ylabel, xlabel, opcao=' '):

plota_pivot_table(sinasc, 'IDADEMAE', 'DTNASC', 'mean', 'Média Idade da Mãe', 'Data de Nascimento' )
plt.savefig(os.path.join(diretorio_salvamento, max_data, 'media_idade_mae_por_data.png'))

plota_pivot_table(sinasc, 'PESO', 'ESCMAE', 'median', 'Peso Bebê', 'Escolaridade Mãe', 'sort' )
plt.savefig(os.path.join(diretorio_salvamento, max_data, 'peso_por_escolaridade_mae.png'))

plota_pivot_table(sinasc, 'IDADEMAE', ['DTNASC', 'SEXO'], 'mean', 'Média Idade Mãe', 'Data de Nascimento', 'unstack' )
plt.savefig(os.path.join(diretorio_salvamento, max_data, 'media_idade_mae_por_sexo.png'))

plota_pivot_table(sinasc, 'PESO', ['DTNASC', 'SEXO'], 'mean', 'média peso bebê', 'data de nascimento', 'unstack')
plt.savefig(os.path.join(diretorio_salvamento, max_data, 'media_peso_bebe_por_sexo.png'))

plota_pivot_table(sinasc, 'PESO', 'ESCMAE', 'median', 'PESO mediano', 'escolaridade mãe', 'sort')
plt.savefig(os.path.join(diretorio_salvamento, max_data, 'peso_mediano_por_escolaridade_mae.png'))

plota_pivot_table(sinasc, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio', 'gestacao', 'sort')
plt.savefig(os.path.join(diretorio_salvamento, max_data, 'apgar1_medio_por_gestacao.png'))
            
```
